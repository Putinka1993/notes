Когда стоит вычислять столбцы в PQ, а когда на DAX?

В лекции Вы познакомились с вычислением столбцов на DAX. А до этого делали их вычисление на языке М в Power Query.

И с точки зрения бизнес-пользователя разницы вроде как и нет – получите тот же результат. Ну и DAX бизнес-пользователю часто будет выглядеть предпочтительнее (Вы это далее увидите), так как это язык функций и работает сходно с Excel. А сделать сложные вычисления в Power Query не зная языка M бывает проблематично.

Т.е., главный критерий выбора М или DAX на уровне бизнес-пользователя сводится к ответу на вопрос: а знает ли пользователь язык М (который сложнее DAX), чтобы сделать любые вычисления в PQ?

Но все-таки считают, что и бизнес-пользователям также следует знать, когда лучше применять М, а когда DAX.

Если речь идет об ETL (извлечение, преобразование и загрузка) этапе работе с данными – то предпочтительнее все делать в Power Query.

Общее правило\логика (рекомендуемая Microsoft) - любые преобразования и вычисления надо перекладывать как можно ближе к источнику. Т.е., в идеале, чтобы все вообще делал источник (проводить вычисления в базе данных) - если нет, то тогда М\PQ, и только потом уже DAX.

Да и с технической стороны вопросы это наиболее логичный шаг: ведь если вся обработка на источнике данных или PQ - то визуализации быстрее "летают" и отрисовываются. Ну и памяти «жрет» меньше. Например, 12 млн строк вычисленных на DAX занимают почти в 4 раза больше памяти, чем такое же вычисление в PQ\M.

Вообще с технической т.з. надо принимать во внимание и обновления данных, и компрессию, и время обработки.

И мой ответ на вопрос «когда лучше вычислять столбец в PQ, а когда на DAX?» совпадает с правилом Microsoft (т.е., вычислять столбцы в PQ или вообще в источнике), кроме 2-х случаев:

- когда надо вычислить столбец из столбцов разных таблиц (на DAX Вы можете это сделать, к примеру, через функцию RELATED)

-  если для вычисления значений столбца используются разные строки таблицы (редкие случаи «закрученных» вычислений)

Во всех остальных случаях (когда вычисление столбца идет по значениям других столбцов в пределах одной таблицы) делайте вычисления в PQ. Но если не знаете, как сделать вычисление в PQ (пользовательских функционалов недостаточно, а знаний М для нужного вычисления у Вас нет) - то используете DAX.
