# Список тренировочных задач

В данном разделе курса мы представим большой список тренировочных задач по каждой из пройденных тем. Мы предлагаем вам реализовать каждую задачу как средствами библиотеки pandas, так и на SQL. Наши варианты решения задач вы сможете найти в этом же разделе в конце. Каждая задача имеет свой порядковый номер, по которому будет удобно искать соответствующий ответ.

### Map операции

#### 1.1

В таблице order_details есть поле discount_price, оно хранит данные о стоимости на позицию товара в чеке с учетом скидки, при этом в одной позиции может быть несколько товаров, следовательно, столбец discount_price содержит итоговую стоимость всех товаров в позиции. Получите новый вычислямый столбец, котором будет расчитана цена за 1 товар с учетом скидки.

#### 1.2

Создайте новый вычисляемый столбец в таблице customers, который будет хранить данные о диапазонах возрастов клиентов. Возраст расчитайте относительно даты, в которую вы решаете данную задачу. Используйте следующие рамки для расчета диапазонов:

 - 18 - 24 года
 - 25 - 34 года
 - 35 - 44 года
 - 45 - 54 года
 - 55 и более

#### 1.3

В таблице сотрудников магазина salesmen создайте два вычисляемых поля, в одном должно храниться имя сотрудника, в другом - фамилия.  

#### 1.4

В таблице orders создайте новое поле, хранящее информацию о годе и номере квартала каждого заказа.

#### 1.5

В таблице order_details в новом столбце рассчитайте сумму скидки в процентах на каждую позицию товаров, результат округлите с точность до 1 знака после запятой.

#### 1.6

Создайте новый вычисляемый столбец в таблице products с классификацией товаров в зависимости от их цены. Использйте следующие диапазоны:

 - цена до 10 000 (невключительно) - класс "low price"
 - цена от 10 000 (включительно) до 100 000 (невключительно) - класс "medium price"
 - цена от 100 000 (включительно) и выше - класс "high price"

### Оконные функции, накопительный итог

#### 7.1

В таблице products создайте новый столбец, показывающий в каждой строке общее количество уникальных товаров, продающихся в сети магазинов.

#### 7.2

В таблице products напротив каждого бренда проставьте среднюю цену составляющих его товаров, гранулярность таблицы products не должна измениться.

#### 7.3

Теперь добавьте разбивку также по категории, иными словами, рассчитайте среднюю цену товаров, входящих в каждый бренд каждой категории. Гранулярность таблицы products не должна измениться.

#### 7.4

В таблице orders напротив каждого клиента получите общее количество его заказов за всю историю, при этом гранулярность таблицы не должна измениться.

#### 7.5

По каждому сотруднику магазинов в таблице orders рассчитайте максимальную сумму чека с учетом скидки среди когда-либо оформленных заказов, гранулярность таблицы orders не должна измениться.

#### 7.6

По каждому клиенту в таблице orders расчитайте накопительную сумму по его заказам (с учетом скидки), отсортировав даты покупок по возрастанию.

#### 7.7

В таблице stores проранжируйте магазины в зависимости от их площади. Самый большой по площади магазин должен получить ранг 1, если площадь у магазинов совпадает, то они должны получить одинаковые ранги, а последующие магазины ранг по порядку (без перескакивания).

#### 7.8

В таблице orders проранжируйте заказы для каждого клиента, отсортировав покупки по возрастанию даты. Иными словами, первый заказ должен получить ранг 1, если даты заказов совпадут, тогда высший ранг (с меньшим по модулю числом) должен получить заказ с большей суммой чека.

#### 7.9

Для каждого клиента в таблице orders получите дату его предыдущего и последующего заказа.
