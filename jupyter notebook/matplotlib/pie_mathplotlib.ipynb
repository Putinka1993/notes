```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import AutoMinorLocator
from matplotlib.gridspec import GridSpec
from sklearn.datasets import load_iris
```

```python
df_values = [['Цех №1', 300],
             ['Цех №2', 210],
             ['Цех №3', 157],
             ['Цех №4', 180],
             ['Цех №5', 98]]
df_columns = ['Цех', 'Произведено деталей, шт']
df = pd.DataFrame(df_values, columns=df_columns)

display(df)
```

```python
plt.pie(df['Произведено деталей, шт'],
       labels=df['Цех'],
       wedgeprops={'edgecolor': 'white', # цвет отделящей линии 
                  'linewidth': 3, # толщина линии между секциями 
                  'width': 0.7}, # диаметр внутренней окружности
       shadow=True) # тень
plt.show()
```

```python
explode = [0, 0, 0.2, 0, 0.1]
plt.pie(df['Произведено деталей, шт'],
       labels=df['Цех'],
       wedgeprops={'edgecolor': 'white', # цвет отделящей линии 
                  'linewidth': 3, # толщина линии между секциями 
                  'width': 0.9}, # диаметр внутренней окружности
       shadow=True,
       explode=explode) # тень
plt.show()
```

```python
explode = [0, 0, 0, 0, 0.2]
plt.pie(df['Произведено деталей, шт'],
       labels=df['Цех'],
       wedgeprops={'edgecolor': 'white', # цвет отделящей линии 
                  'linewidth': 3, # толщина линии между секциями 
                  'width': 0.9}, # диаметр внутренней окружности
       shadow=True, # тень
       explode=explode, # Параметру explode передаётся список, в котором на месте, соответствующем выдвигаемой доли, должно быть число, отличное от нуля
       startangle=90) # развернуть круг на N градусов
plt.show()
```

```python
colors = ['#B1B176', '#DBA4F6', '#B7DC82', '#A4F6CB', '#C5CCAD']
explode = [0, 0, 0, 0, 0.2]
plt.pie(df['Произведено деталей, шт'],
       labels=df['Цех'],
       wedgeprops={'edgecolor': 'white', # цвет отделящей линии 
                  'linewidth': 3, # толщина линии между секциями 
                  'width': 0.9}, # диаметр внутренней окружности
       shadow=True, # тень
       explode=explode, # Параметру explode передаётся список, в котором на месте, соответствующем выдвигаемой доли, должно быть число, отличное от нуля
       startangle=90,# развернуть круг на N градусов
       colors=colors) # задать цвета
plt.show()
```
