```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import AutoMinorLocator
from matplotlib.gridspec import GridSpec
from sklearn.datasets import load_iris
```

```python
df_values = [['2019', 7238, 7239],
             ['2020', 7497, 7840],
             ['2021', 8997, 8410],
             ['2022', 9472, 9107]]
df_columns = ['Год', 'Торговые площади Магнит, тыс. кв.м.',
              'Торговые площади X5Group, тыс. кв.м.']
df = pd.DataFrame(df_values, columns=df_columns)

display(df)
```

```python
plt.bar(df['Год'], df['Торговые площади X5Group, тыс. кв.м.'],
       width=0.3, color='purple', alpha=0.5)

plt.title('Торговые площади X5Group')
plt.xlabel('YEAR')
plt.ylabel('area thousand square killometers')

plt.show()
```

```python
plt.barh(df['Год'], df['Торговые площади X5Group, тыс. кв.м.'],
        height=0.5, color='purple', alpha=0.5)

plt.title('Торговые площади X5Group')
plt.xlabel('YEAR')
plt.ylabel('area thousand square killometers')

plt.show()
```

```python
ax = plt.bar(df['Год'], df['Торговые площади X5Group, тыс. кв.м.'],
       width=0.3, color='purple', alpha=0.5)

plt.title('Торговые площади X5Group')
plt.xlabel('YEAR')
plt.ylabel('area thousand square killometers')

plt.bar_label(ax)

plt.show()
```

```python
plt.figure(figsize=(8, 6))

w = 0.1
x = df.index.values

plt.bar(x-w, df['Торговые площади Магнит, тыс. кв.м.'],
        width=w*2, color='green', alpha=0.5)

plt.bar(x+w, df['Торговые площади X5Group, тыс. кв.м.'], 
        width=w*2, color='red', alpha=0.5)

plt.xticks(ticks=x, labels=df['Год'])

plt.legend(['Магнит', 'X5Group'])
plt.title('Торговые площади')
plt.xlabel('Год')
plt.ylabel('Торговые площади, тыс. кв.м.')

plt.show()
```
