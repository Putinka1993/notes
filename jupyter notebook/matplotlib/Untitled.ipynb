```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-np.pi, np.pi, 20)
cos_x = np.cos(x) 
sin_x = np.sin(x) 
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2) 
ax2 = ax.plot(x, sin_x, marker='*', c='r', linewidth=2)

ax.tick_params(labelrotation=45)

plt.show()
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2) 
ax2 = ax.plot(x, sin_x, marker='*', c='r', linewidth=2)

ax.tick_params(axis='x', labelrotation=90)

plt.show()
```
