```python
values = {'x': [2, 4, 8, 12, 15, 20, 25], 'y': [0, 5, 4, 11, 12, 10, 9]}
df = pd.DataFrame(values)
```

```python
plt.plot(df['x'], df['y'], color='brown', linewidth=2, alpha=0.7, linestyle='-')
plt.plot(df['x'], df['y']+2, color='purple', linewidth=2, alpha=0.7, linestyle='--')
plt.plot(df['x'], df['y']+4, color='royalblue', linewidth=2, alpha=0.7, linestyle='-.')
plt.plot(df['x'], df['y']+6, color='coral', linewidth=2, alpha=0.7, linestyle=':')


plt.grid(color='grey', linestyle='--', linewidth=0.5)
plt.title('График зависимости признаков x и y', fontsize=14)
plt.xlabel(df.columns[0], fontsize=15)
plt.ylabel(df.columns[1], fontsize=15, rotation=1)
plt
```

```python
plt.plot(df['x'], df['y'], color='black', linewidth=3, marker='o', markerfacecolor='blue', markersize=10)
plt
```

```python
values = {'x1': [2, 4, 8, 12, 15, 20, 25], 
          'x2': [1, 3, 7, 10, 13, 18, 22],
          'y1': [0, 5, 4, 11, 12, 10, 9],
          'y2': [3, 10, 15, 6, 5, 2, 1]}

df = pd.DataFrame(values)
```

```python
plt.plot(df['x1'], df['y1'], 
        color='yellow', linewidth=3, marker='o')
plt.plot(df['x2'], df['y2'],
        color='red', linewidth=3, marker='o')
plt
```

```python
plt.plot(df['x1'], df['y1'],
        linewidth=3, color='black', marker='o')
plt.plot(df['x2'], df['y2'],
        linewidth=3, color='blue', marker='o')
plt.legend(['x1_y1', 'x2_y2'], loc='upper left')
```

```python
plt.plot(df['x1'], df['y1'],
        linewidth=3, marker='o')
plt.plot(df['x2'], df['y2'],
        linewidth=3, marker='o')
plt.style.use('ggplot')

```
