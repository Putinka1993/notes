```python
plt.plot(df['x1'], df['y1'], 
        color='yellow', linewidth=3, marker='o')
plt.plot(df['x2'], df['y2'],
        color='red', linewidth=3, marker='o')
plt
```

```python
plt.plot(df['x1'], df['y1'],
        linewidth=3, color='black', marker='o')
plt.plot(df['x2'], df['y2'],
        linewidth=3, color='blue', marker='o')
plt.legend(['x1_y1', 'x2_y2'], loc='upper left')
```

```python
x = np.linspace(0, 10, 5)
y1 = x**2
y2 = x**3
y3 = x**0.5
y4 = np.sin(x**2)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r')
plt.title('y1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g')
plt.title('y2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'y')
plt.title('y3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'c')
plt.title('y4')

plt.style.use('default')

plt.suptitle('Графики')

plt.tight_layout()

plt.show()
```
