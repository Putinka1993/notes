```python
values = {'x': [2, 4, 8, 12, 15, 20, 25], 'y': [0, 5, 4, 11, 12, 10, 9]}
df = pd.DataFrame(values)
```

```python
plt.plot(df['x'], df['y'])
plt.grid(color='grey', linestyle='--', linewidth=0.5)
plt.title('График зависимости признаков X и Y', fontsize=14)
plt.xlabel(df.columns[0], fontsize=14, color='purple')
plt.ylabel(df.columns[1], fontsize=14, color='purple')

plt.xticks([0, 10, 20, 30])
plt.yticks([0, 5, 10, 15])

plt.show()
```

```python
plt.plot(df['x'], df['y'])
plt.grid(color='grey', linestyle='--', linewidth=0.5)
plt.title('График зависимости признаков X и Y', fontsize=14)
plt.xlabel(df.columns[0], fontsize=14, color='purple')
plt.ylabel(df.columns[1], fontsize=14, color='purple')

plt.xticks(range(0, 26), rotation=90)
plt.yticks([0, 1, 2, 3, 4, 5, 10], rotation=15)

plt.show()
```
