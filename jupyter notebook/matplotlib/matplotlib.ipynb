```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
values = [['Австралия', 25], ['Азия', 4540], ['Африка', 1280],
          ['Европа', 742], ['Северная Америка', 543], ['Южная Америка', 428]]
columns = ['Часть света', 'Численность в 2018 г., млн чел.']
df = pd.DataFrame(values, columns=columns)

df
```

```python
plt.plot(df['Часть света'], df['Численность в 2018 г., млн чел.'])
```

```python
values = {'x': [2, 4, 8, 12, 15, 20, 25], 'y': [0, 5, 4, 11, 12, 10, 9]}
df = pd.DataFrame(values)

plt.plot(df['x'], df['y'])
plt.grid(color='pink', linestyle='--', linewidth=2, alpha=1, axis='both')
```

```python
plt.plot(df['x'], df['y'])
plt.grid(color='grey', linestyle='--', linewidth=0.5)

plt.title('График зависимости x и y', fontsize=15, color='green', alpha=0.7)

plt.xlabel(df.columns[0], fontsize=14, color='purple', labelpad=15)
plt.ylabel(df.columns[1], fontsize=14, color='purple', labelpad=15, rotation=0)

plt.show()
```
