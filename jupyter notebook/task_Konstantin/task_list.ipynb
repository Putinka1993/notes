```python
sql = '''
select
	c.birthday,
	extract(year from age(current_date, c.birthday)),
	case
		when extract(year from age(current_date, c.birthday)) between 18 and 24 then '18 - 24 года'
		when extract(year from age(current_date, c.birthday)) between 25 and 34 then '25 - 34 года'
		when extract(year from age(current_date, c.birthday)) between 35 and 44 then '35 - 44 года'
		when extract(year from age(current_date, c.birthday)) between 45 and 54 then '45 - 54 года'
		else '55 и более'
	end as age_group
from sales.customers c
'''
```

```python
# pandas
def age_group(age):
    if 18 <= age <= 24:
        return '18 - 24 года'
    elif 25 <= age <= 34:
        return '25 - 34 года'
    elif 35 <= age <= 44:
        return '35 - 44 года'
    elif 45 <= age <= 54:
        return '45 - 54 года'
    else:
        return '55 и более'
        
customers['birthday'] = pd.to_datetime(customers['birthday'])
customers["age"] = customers['birthday'].apply(
    lambda x: pd.Timestamp.today().year - x.year - ((pd.Timestamp.today().month, pd.Timestamp.today().day) < (x.month, x.day))
)
customers["age_group"] = customers['age'].apply(age_group)
customers[["birthday", "age_group"]].head()
```

#### 7.6

По каждому клиенту в таблице orders расчитайте накопительную сумму чеков по его заказам (с учетом скидки), отсортировав даты покупок по возрастанию.

```python
sql = '''

select 
		o.order_date,
		o.customer_id,
		o.order_discount_cost,
		sum(o.order_discount_cost) over(partition by o.customer_id order by o.order_date asc
										rows between unbounded preceding and current row) as cum_sum_by_customers
from sales.orders o 

'''
```

```python
orders = pd.read_csv("orders.csv", sep = '|') 

# Преобразуем дату в формат datetime
orders['order_date'] = pd.to_datetime(orders['order_date'])

# Сортируем по клиенту и дате заказа
orders = orders.sort_values(by=['customer_id', 'order_date'])

# Вычисляем накопительную сумму чеков по каждому клиенту
orders['cum_sum_by_customers'] = orders.groupby('customer_id')['order_discount_cost'].cumsum()

# Вывод результата
orders[['order_date', 'customer_id', 'order_discount_cost', 'cum_sum_by_customers']].head()
```

#### 7.7

В таблице stores проранжируйте магазины в зависимости от их площади. Самый большой по площади магазин должен получить ранг 1, если площадь у магазинов совпадает, то они должны получить одинаковые ранги, а последующие магазины ранг по порядку (без перескакивания).

#### 7.8

В таблице orders проранжируйте заказы для каждого клиента, отсортировав покупки по возрастанию даты. Иными словами, первый заказ должен получить ранг 1, если даты заказов совпадут, тогда высший ранг (с меньшим по модулю числом) должен получить заказ с большей суммой чека.

```python
sql = '''

select 
		o.order_date,
		o.order_discount_cost,
		o.customer_id,
		row_number() over(partition by o.customer_id order by o.order_date asc, o.order_discount_cost desc) as order_rank
from sales.orders o

'''
```

```python
orders = pd.read_csv("orders.csv", sep = '|')  

# Преобразуем дату в формат datetime
orders['order_date'] = pd.to_datetime(orders['order_date'])

# Сортируем по клиенту, дате заказа (по возрастанию) и сумме чека (по убыванию)
orders = orders.sort_values(by=['customer_id', 'order_date', 'order_discount_cost'], ascending=[True, True, False])

# Присваиваем ранг каждому заказу клиента
orders['order_rank'] = orders.groupby('customer_id').cumcount() + 1

# Вывод результата
orders[['order_date', 'order_discount_cost', 'customer_id', 'order_rank']].head()
```

Для каждого клиента в таблице orders получите дату его предыдущего и последующего заказа (в двух разных столбцах).

```python
sql = '''

select 
		o.customer_id,
		o.order_date,
		lag(o.order_date) over(partition by o.customer_id order by o.order_date asc) as lag_date,
		lead(o.order_date) over(partition by o.customer_id order by o.order_date asc) as lead_date
from sales.orders o 

'''
```

```python
orders = pd.read_csv("orders.csv", sep = '|')  

# Преобразуем дату в формат datetime
orders['order_date'] = pd.to_datetime(orders['order_date'])

# Сортируем по клиенту и дате заказа
orders = orders.sort_values(by=['customer_id', 'order_date'])

# Вычисляем дату предыдущего и следующего заказа
orders['lag_date'] = orders.groupby('customer_id')['order_date'].shift(1)
orders['lead_date'] = orders.groupby('customer_id')['order_date'].shift(-1)

# Вывод результата
orders[['customer_id', 'order_date', 'lag_date', 'lead_date']].head()
```
