```python
customers = pd.read_csv('/Users/vladislavlipkin/Documents/Konstantin/customers.csv', sep='|')
```

```python
customers.head()
```

```python
# pandas

customers['birthday'] = pd.to_datetime(customers['birthday'])

def get_range_age(row):
    if 18 <= (pd.Timestamp.now() - row['birthday']).days /  365.25 <= 24:
        return '18 - 24'
    elif 25 <= (pd.Timestamp.now() - row['birthday']).days /  365.25 <= 34:
        return '25 - 34'
    elif 35 <= (pd.Timestamp.now() - row['birthday']).days /  365.25 <= 44:
        return '35 - 44'
    elif 45 <= (pd.Timestamp.now() - row['birthday']).days /  365.25 <= 54:
        return '45 - 54'
    return '55 and lot'
    
customers['age_range'] = customers.apply(get_range_age, axis=1)

# SQL

# with age_table as (
# 	select
# 		customer_name ,
# 		birthday ,
# 		extract(year from age(birthday)) as AGE
# 	from
# 		sales.customers
# )
# select
# 	customer_name ,
# 	birthday ,
# 	AGE ,
# 	case 
# 		when AGE between 18 and 24 then '35 - 44'
# 		when AGE between 25 and 34 then '35 - 44'
# 		when AGE between 35 and 44 then '35 - 44'
# 		when AGE between 45 and 54 then '35 - 44'
# 		when AGE > 55 then '55 и более'
# 	end as age_group
# from
# 	age_table
```

```python
# 1.6
# Создайте новый вычисляемый столбец в таблице products с классификацией товаров в зависимости от их цены. Использйте следующие диапазоны:

# цена до 10 000 (невключительно) - класс "low price"
# цена от 10 000 (включительно) до 100 000 (невключительно) - класс "medium price"
# цена от 100 000 (включительно) и выше - класс "high price"
```

```python
# pandas
products = pd.read_csv(r'C:\Users\Incognitus\Documents\Daniel_and_Konstantin_csv\products.csv', sep='|')

products.head()
def level_price(row):
    if row['price'] < 10000:
        return 'Low price'
    elif 10000 < row['price'] < 100000:
        return 'Medium price'
    return 'High price'

products['class_products'] = products.apply(level_price, axis=1)
products.head()

# sql
# select
# 	price ,
# 	case
# 		when price < 10000 then 'low price'
# 		when price >= 10000 then 'medium price'
# 		when price >= 100000 then 'high price'
# 	end as cassification
# from
# 	sales.products;
```
